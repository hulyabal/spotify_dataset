### 1. Valence

It is a measure of how positive or negative a piece of music feels.

```python
df['valence'].nunique()
```

```python
df['valence'].describe()
```

```python
df['valence']
```

```python
def histogram_plot(data, column_name, xlabel, title):
    plt.figure(figsize=(6, 4))
    plt.hist(data[column_name], bins=50, edgecolor='black')
    plt.xlabel(xlabel, fontsize=8, fontweight='bold')
    plt.ylabel('Frequency', fontsize=8, fontweight='bold')
    plt.title(title, fontsize=10, fontweight='bold')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.gca().set_facecolor('#f9f9f9')
    plt.tight_layout()
    plt.show()
```

```python
histogram_plot(df, 'valence', 'Valence', 'Distribution of Valence Scores')
```

```python
valence_data = df['valence']

high_count = valence_data[valence_data >= 0.5].count()
low_count = valence_data[valence_data < 0.5].count()
total_count = high_count + low_count

high_percentage = (high_count / total_count) * 100
low_percentage = (low_count / total_count) * 100

labels = ['High Valence (>= 0.5)', 'Low Valence (< 0.5)']
sizes = [high_percentage, low_percentage]
colors = ['skyblue', 'salmon']

plt.figure(figsize=(4, 4))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})

plt.title('Distribution of High and Low Valence Scores', fontsize=10, fontweight='bold')

plt.show()
```

### 3. Acousticness

As it nears 0, electronic sounds dominate, whereas approaching 1 indicates a more natural sound.

```python
df['acousticness'].describe()
```

```python
histogram_plot(df, 'acousticness', 'Acousticness', 'Distribution of Acousticness Scores')
```

```python
correlation_matrix = df.corr()
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".1f", mask=mask)
plt.title('Correlation Matrix', fontsize=12, fontweight='bold')
plt.show()
```

```python
high_corr = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if correlation_matrix.iloc[i, j] > 0.70:
            high_corr.append((correlation_matrix.columns[i], correlation_matrix.columns[j], correlation_matrix.iloc[i, j]))

print("Features with correlation higher than 0.70:")
for feature1, feature2, corr_value in high_corr:
    print(f"{feature1} and {feature2}: {corr_value:.2f}")
```

The high correlation between "energy" and "loudness" (0.78) indicates that energetic songs are generally louder. The strong correlation between "year" and "popularity" (0.86) reflects that songs released in certain years tend to be more popular.

```python
acousticness_by_year = df.groupby('year')['acousticness'].mean()

plt.plot(acousticness_by_year.index, acousticness_by_year.values, marker='o', color='blue')
plt.xlabel('Year',fontsize=8)
plt.ylabel('Mean Acousticness',fontsize=8)
plt.title('Mean Acousticness by Year',fontsize=10, fontweight='bold')
plt.xticks(np.arange(min(df['year']), max(df['year'])+1, 10))
plt.grid(True)
plt.show()
```

Over time, songs have become less natural, indicating a shift from natural sounds in the music industry.

```python
acousticness_ts = df.groupby('year')['acousticness'].mean()
energy_ts = df.groupby('year')['energy'].mean()

plt.figure(figsize=(10, 4))
plt.plot(acousticness_ts.index, acousticness_ts.values, label='Acousticness', color='blue')
plt.plot(energy_ts.index, energy_ts.values, label='Energy', color='red')
plt.xlabel('Year')
plt.ylabel('Level')
plt.title('Mean Acousticness and Energy Over Time',fontsize=10, fontweight='bold')
plt.legend()
plt.grid(True)
plt.show()
```

```python
popular_artists = df.groupby('artists')['popularity'].mean().sort_values(ascending=False).head(10)

artist_yearly_count = df.groupby(['artists', 'year']).size().unstack(fill_value=0)
top_artists_yearly_count = artist_yearly_count.loc[popular_artists.index]

sns.set(style="whitegrid")
palette = sns.color_palette("bright", len(top_artists_yearly_count))

plt.figure(figsize=(14, 8))
for idx, artist in enumerate(top_artists_yearly_count.index):
    plt.plot(top_artists_yearly_count.columns, top_artists_yearly_count.loc[artist], 
             label=artist, linewidth=2, color=palette[idx], marker='o', markersize=6, linestyle='--' if idx % 2 == 0 else '-')


plt.title('Yearly Song Production of Top 10 Artists', fontsize=12, fontweight='bold')
plt.xlabel('Year', fontsize=10)
plt.ylabel('Number of Songs', fontsize=10)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)
plt.ylim(0, 5) 
plt.yticks(range(0, 6, 1)) 
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Artists', fontsize=10, title_fontsize=10, loc='upper left')

plt.tight_layout()
plt.show()
```

```python
artist_decade_count = df.groupby(['artists', 'decade']).size().unstack(fill_value=0)

top_artists_by_decade = {}
for decade in artist_decade_count.columns:
    top_artists = artist_decade_count[decade].nlargest(5).index.tolist()
    top_artists_by_decade[decade] = artist_decade_count.loc[top_artists, decade]

num_decades = len(top_artists_by_decade)
rows = (num_decades + 1) // 2

fig, axes = plt.subplots(rows, 2, figsize=(20, 5 * rows))
fig.subplots_adjust(hspace=0.4)

palette = sns.color_palette("husl", 5)

for i, (decade, data) in enumerate(top_artists_by_decade.items()):
    row = i // 2
    col = i % 2
    ax = axes[row, col]
    
    sns.barplot(x=data.values, y=data.index, ax=ax, palette=palette)
    
    for p in ax.patches:
        ax.annotate(f'{int(p.get_width())}', (p.get_width(), p.get_y() + p.get_height() / 2),
                    ha='left', va='center', fontsize=15, fontweight='bold', color='black', xytext=(5, 0),
                    textcoords='offset points')
    
    ax.set_title(f'Top 5 Artists in {decade}s', fontsize=16, fontweight='bold')
    ax.set_xlabel('Number of Songs', fontsize=14)
    ax.set_ylabel('Artists', fontsize=14)
    ax.tick_params(axis='both', which='major', labelsize=14)
    ax.yaxis.set_tick_params(labelsize=16, width=3)  
    ax.grid(True, linestyle='--', alpha=0.6)

if num_decades % 2 != 0:
    fig.delaxes(axes[-1, -1])

plt.tight_layout()
plt.show()
```

```python
popular_artists = df.groupby('artists')['popularity'].mean().sort_values(ascending=False).head(10)

df['duration_min'] = df['duration_ms'] / (1000 * 60)
artist_avg_duration = df.groupby('artists')['duration_min'].mean().loc[popular_artists.index]

artist_avg_duration_sorted = artist_avg_duration.sort_values(ascending=False)

plt.figure(figsize=(10, 4))
sns.barplot(x=artist_avg_duration_sorted.index, y=artist_avg_duration_sorted.values, palette='coolwarm')
plt.title('Average Song Duration per Minute for Top 10 Artists', fontsize=10, fontweight='bold')
plt.xlabel('Artists', fontsize=10)
plt.ylabel('Average Song Duration (minutes)', fontsize=10)
plt.xticks(rotation=45, fontsize=10,fontweight='bold')
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
def boxplot_plot(data, column_name, xlabel, title, figsize=(6, 4), color='skyblue', width=0.3, linewidth=1.5, fliersize=3):
    plt.figure(figsize=figsize)
    sns.boxplot(x=column_name, data=data, color=color, width=width, linewidth=linewidth, fliersize=fliersize)
    plt.xlabel(xlabel, fontsize=8, fontweight='bold')
    plt.ylabel('Value', fontsize=8, fontweight='bold')
    plt.title(title, fontsize=10, fontweight='bold')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.gca().set_facecolor('#f9f9f9')
    plt.tight_layout()
    plt.show()
```

```python
def plot_energy_trend(df, feature):
    """
    Plot the trend of a given feature over the years with a line plot and a regression line.
    
    Parameters:
    df (pd.DataFrame): DataFrame containing the data.
    feature (str): The name of the feature to plot.
    """
    plt.figure(figsize=(10, 4))

    sns.lineplot(x='year', y=feature, data=df, estimator='mean', ci=None, color='navy', linewidth=3)

    sns.regplot(x='year', y=feature, data=df, scatter=False, color='red', line_kws={'linewidth': 2})

    plt.title(f'Change in {feature.capitalize()} Score Over Years', fontsize=10, fontweight='bold')
    plt.xlabel('Year', fontsize=8)
    plt.ylabel(f'Mean {feature.capitalize()}', fontsize=8)
    plt.xticks(rotation=45, fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.6)

    plt.show()
```

```python
plot_energy_trend(df, 'danceability')
```
